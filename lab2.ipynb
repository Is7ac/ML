```python
import matplotlib.pyplot as plt
```

```python

# Data
x1 = [1, 5, 7, 2]
x2 = [1, 2, 2, 3]
y = [2, 4, 2, 4]

# Plotting the data
plt.scatter(x1, y, color='red', label='X1')
plt.scatter(x2, y, color='blue', label='X2')
plt.xlabel('X1, X2')
plt.ylabel('Y')
plt.title('Scatter Plot of Data')
plt.legend()
plt.grid(True)
plt.show()
```
