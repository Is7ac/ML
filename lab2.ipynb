```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data
x1 = np.array([1, 5, 7, 2])
x2 = np.array([1, 2, 2, 3])
y = np.array([2, 4, 2, 4])

# Plotting the data
plt.scatter(x2, y, color='blue', label='X2')
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('Scatter Plot of Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
theta = np.array([3, 0.5])
learning_rate = 0.25
iterations = 4
```

```python
def hypothesis(x, theta):
    return theta[0] + theta[1] * x
```

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
losses = []
for i in range(iterations):
    # Prediction
    y_pred = hypothesis(x2, theta)

    # Compute gradients
    grad0 = np.mean(y_pred - y)
    grad1 = np.mean((y_pred - y) * x2)

    # Update parameters
    theta[0] -= learning_rate * grad0
    theta[1] -= learning_rate * grad1

    # Compute loss
    loss = mean_squared_error(y, y_pred)
    losses.append(loss)

    # Plot regression model
    plt.scatter(x2, y, color='blue', label='X2')
    plt.plot(x2, hypothesis(x2, theta), color='green', label='Regression Model')
    plt.xlabel('X2')
    plt.ylabel('Y')
    plt.title('Regression Model after Iteration {}'.format(i+1))
    plt.legend()
    plt.grid(True)
    plt.show()

    print("Iteration {}: Theta = {}, Loss = {}".format(i+1, theta, loss))

# Plot loss curve
plt.plot(range(1, iterations+1), losses, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.grid(True)
plt.show()
```

in the initial iteration the loss was 0.955078125 but in the final one the loss became 0.8155279830098152!!
